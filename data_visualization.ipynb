### 4.3 수치형 데이터 시각화

```python
import seaborn as sns

titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

#### 4.3.1 히스토그램(histplot)
- 수치형 데이터의 구간 별 빈도수를 나타내는 그래프

```python
sns.histplot(data = titanic, x = 'age')
```

```python
sns.histplot(data = titanic, x = 'age', bins = 10)
```

```python
sns.histplot(data = titanic, x = 'age', hue = 'alive') # 연령 별 생존여부 중첩
```

```python
sns.histplot(data = titanic, x = 'age', hue = 'alive', multiple = 'stack') # 연령 별 생존여부 누적
```

#### 4.3.2 커널밀도추정 함수 그래프(kdeplot)
- 히스토그램을 곡선으로 연결한 그래프

```python
sns.kdeplot(data = titanic, x = 'age')
```

```python
sns.kdeplot(data = titanic, x = 'age', hue = 'alive', multiple = 'stack')
```

#### 4.3.3 분포도(displot)
- 수치형 데이터 하나의 분포를 나타내는 그래프

```python
sns.displot(data = titanic, x = 'age')
```

```python
sns.displot(data = titanic, x = 'age', kind = 'kde') # kde 그래프
```

```python
sns.displot(data = titanic, x = 'age', kde = True)
```

#### 4.3.4 러그플롯(rugplot)
- 주변 분포(marginal distribution)를 나타내는 그래프

```python
sns.kdeplot(data = titanic, x = 'age')
sns.rugplot(data = titanic, x = 'age')
```

### 4.4 범주형 데이터 시각화
#### 4.4.1 막대그래프(barplot)
- 범주형 데이터 값에 따라 수치형 데이터 값이 어떻게 달라지는지 파악할 때 사용
- 수치형 데이터 평균은 막대 높이로, 신뢰구간은 오차 막대(error bar)로 표현 
    - 원본 데이터 평균이 아닌, 샘플링한 데이터 평균을 구함

```python
sns.barplot(x = 'class', y = 'fare', data = titanic) #평균
# 막대 높이 : 등급 별 평균 운임
# 검은색 세로줄 : 오차 막대(신뢰구간)
```

```python
import numpy as np
sns.barplot(x = 'class', y = 'fare', data = titanic, estimator = np.median) # 중앙값
```

```python
sns.barplot(x = 'class', y = 'fare', data = titanic, estimator = np.max) # 최댓값
```

```python
sns.barplot(x = 'class', y = 'fare', data = titanic, estimator = np.min) # 최솟값
```

#### 4.4.2 포인트플롯(pointplot)
- 범주형 데이터에 따른 수치형 데이터의 평균과 신뢰구간을 점과 선으로 표현

```python
sns.pointplot(x = 'class', y = 'fare', data = titanic)
```

#### 4.4.3 박스플롯(boxplot)
- 5가지 요약 수치(five-number summary) 제공
    - Q1(제1사분위수) : 전체 데이터 중 하위 25%에 해당하는 값
    - Q2(제2사분위수) : 50%에 해당하는 값(중앙값)
    - Q3(제3사분위수) : 상위 25%에 해당하는 값
    - IQR(사분위 범위 수) : Q3 - Q1
    - 최댓값 : Q3 + (1.5 * IQR)
    - 최솟값 : Q1 - (1.5 * IQR)
    - 이상치 : 최댓값보다 큰 값과 최솟값보다 작은 값

```python
sns.boxplot(x = 'class', y = 'age', data = titanic)
```

#### 4.4.4 바이올린플롯(violinplot)
- 박스플롯과 커널밀도추정함수 그래프를 합쳐놓은 그래프

```python
sns.violinplot(x = 'class', y = 'age', data = titanic)
```

```python
sns.violinplot(x = 'class', y = 'age', hue = 'sex', data = titanic, split = True)
```

#### 4.4.5 카운트플롯(countplot)
- 범주형 데이터의 개수를 확인할 때 사용하는 그래프
- 주로 범주형 피처나 범주형 타깃값의 분포가 어떤지 파악하는 용도로 사용

```python
sns.countplot(x = 'class', data = titanic)
```

```python
sns.countplot(y = 'class', data = titanic)
```

#### 4.4.6 파이 그래프(pie)
- 범주형 데이터 별 비율을 알아볼 때 사용하기 좋은 그래프(seaborn에서는 지원X)

```python
import matplotlib.pyplot as plt

x = [10, 60, 30]         # 범주형 데이터 별 파이 그래프의 부채꼴 크기(비율)
labels = ['A', 'B', 'C'] # 범주형 데이터 레이블

plt.pie(x = x, labels = labels, autopct = '%.1f%%')
```

### 4.5 데이터 관계 시각화
#### 4.5.1 히트맵(heatmap)
- 데이터 간 관계를 색상으로 표현한 그래프 

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights_pivot = flights.pivot(index = 'month', columns = 'year', values = 'passengers')
flights_pivot
```

```python
sns.heatmap(data = flights_pivot)
```

#### 4.5.2 라인플롯(lineplot)
- 두 수치형 데이터 사이의 관계를 나타낼 때 사용
- x의 값에 따라 y 값의 평균과 95% 신뢰구간을 나타냄

```python
sns.lineplot(x = 'year', y = 'passengers', data = flights)
```

#### 4.5.3 산점도(scatterplot)
- 두 데이터의 관계를 점으로 표현하는 그래프

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', hue = 'time', data = tips)
```

#### 4.5.4 회귀선을 포함한 산점도 그래프(regplot)
- 산점도와 선형 회귀선을 동시에 그려주는 함수

```python
sns.regplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.regplot(x = 'total_bill', y = 'tip', ci = 99, data = tips) # 신뢰구간 99%로 변경
```
